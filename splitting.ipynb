# SPLITTING DATA

Random sampling adalah teknik pengambilan sampel dimana semua individu dalam populasi, baik secara sendiri-sendiri atau bersama-sama diberi kesempatan yang sama untuk dipilih sebagai anggota sampel.

```python
import random
# membantu mengotomatiskan proses penyalinan dan penghapusan file dan direktori.
from shutil import copyfile

# Fungsi untuk membagi dataset menjadi training dan validation
def train_val_split(source, train, val, train_ratio):
    total_size = len(os.listdir(source))
    if total_size == 0:
        print(f"Folder {source} kosong, tidak ada data yang dibagi.")
        return

    train_size = int(train_ratio * total_size)
    val_size = total_size - train_size

    randomized = random.sample(os.listdir(source), total_size)
    train_files = randomized[0:train_size]
    val_files = randomized[train_size:total_size]

    for i in train_files:
        i_file = os.path.join(source, i)
        destination = os.path.join(train, i)
        copyfile(i_file, destination)

    for i in val_files:
        i_file = os.path.join(source, i)
        destination = os.path.join(val, i)
        copyfile(i_file, destination)


# Jumlah pembagian data training & val
train_ratio = 0.8

# Pembagian Training & val

# kaleng
train_val_split(kaleng_dir, kaleng_train, kaleng_val, train_ratio)
# Botol Plastik
train_val_split(botolPlastik_dir, botolPlastik_train, botolPlastik_val, train_ratio)
# Bungkus Plastik
train_val_split(bungkusPlastik_dir, bungkusPlastik_train, bungkusPlastik_val, train_ratio)
# Cup Gelas
train_val_split(cupGelas_dir, cupGelas_train, cupGelas_val, train_ratio)
# Galon
train_val_split(galon_dir, galon_train, galon_val, train_ratio)
# Kaca
train_val_split(kaca_dir, kaca_train, kaca_val, train_ratio)
# Kertas
train_val_split(kertas_dir, kertas_train, kertas_val, train_ratio)
# Sampah Ban
train_val_split(ban_dir, ban_train, ban_val, train_ratio)
# Sampah Organik
train_val_split(organik_dir, organik_train, organik_val, train_ratio)
# Kardus
train_val_split(kardus_dir, kardus_train, kardus_val, train_ratio)
```
